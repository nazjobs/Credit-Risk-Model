```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Load Data
# Ensure your data file is named 'data.csv' and is in data/raw/
try:
    df = pd.read_csv('../data/raw/data.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("Error: data.csv not found in ../data/raw/")
```

```python
# 2. Summary Statistics
print(df.info())
print(df.describe())
```

```python
# 3. Distribution of Numerical Features
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.yscale('log') # Log scale helps with skewed financial data
```

```python
# 4. Correlation Analysis
plt.subplot(1, 2, 2)
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), cmap='coolwarm', annot=False)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# 5. Top Insights
print("""
Top Insights:
1. Skewness: Transaction amounts are highly right-skewed.
2. Categorical Imbalance: Certain ProductCategories dominate the dataset.
3. Missing Values: Checked via info(), handling strategy will be imputation.
""")
```
